## Sinal impulso unitário

É o sinal mais simples que existe em tempio discreto. Para representá-lo, usamos a letra $\delta$, e é definido da seguinte maneira:

$$\delta[n] = \left\{\begin{array}{cc}1& ,n = 0 \\ 0& ,c.c \end{array}\right. $$

O seu gráfico é o seguinte:


```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10,11)
x = np.zeros_like(n)
x[n==0] = 1

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x');
```

Para obter um impulso unitário em outra posição que não $n=0$, basta atrasarmos (ou adiantarmos) o sinal $\delta[n]$. Se queremos que o sinal seja 1 em $n=k$, basta fazermos $\delta[n-k]$.

Por exemplo, se queremos um impulso unitário em $n=2$, fazemos $\delta[n-2]$.

```python

n = np.arange(-10,11)
x = np.zeros_like(n)
x[n==2] = 1

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x');
```

## Sinal degrau

Outro sinal que usaremos bastante é o sinal degrau de tempo discreto. Esse sinal é sempre igual a 1 para $n\geq 0$. Para esse sinal usaremos a letra $u$:

$$u[n] =   \left\{\begin{array}{cc}1& ,n \geq 0 \\ 0& ,c.c \end{array}\right. $$

O seu gráfico é mostrado abaixo: 

```python
n = np.arange(-10,11)
x = np.zeros_like(n)
x[n>=0] = 1

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x');
```

Os sinais degrau e impulso unitário em tempo discreto estão bastante relacionados. O sinal impulso unitário é o sinal degrau subtraido do mesmo sinal degrau atrasado de 1 passo.

$$\delta[n] = u[n] - u[n-1]$$
